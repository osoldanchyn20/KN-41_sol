### **Визначення однорідного масиву**

```python
lst = [ [3.4, 8.7, 9.9], 
        [1.1, -7.8, -0.7],
        [4.1, 12.3, 4.8] ]

A = np.array(lst, dtype=np.int8) # Створюємо масив типу int8, що перетворює всі елементи до цілого числа в межах від -128 до 127

print(f"Отриманий масив має вигляд:\n{repr(A)}") # Використовуємо repr для відображення строкового подання масиву

B = np.array(lst, dtype=np.float16) # Створюємо масив типу float16, що зберігає числа з плаваючою точкою з меншою точністю

print(f"Отриманий масив має вигляд:\n{repr(B)}")# Використовуємо repr для більш детального подання об'єкта

 

homework("Згадуючи ООП, що робить функція repr?")
homework("У чому різниця між масивами A i B?")

# перше завдання
print("Функція repr повертає офіційне строкове подання об'єкта, яке розробники можуть використовувати для аналізу структури об'єкта.\n")

# друге завдання
print("Різниця між масивами A і B полягає у їхньому типі даних:")
print("Масив A має тип int8, що зберігає цілі числа у діапазоні від -128 до 127, тому дробові частини втрачаються.")
print("Масив B має тип float16, який зберігає числа з плаваючою точкою, але з меншою точністю, ніж float64.")
```

### **Задаємо свій тип даних**

**можливо створити свій набір даних задавши назву та типи полів. Кожне поле повинне бути заданим як tuple який містить імя та тип (числовий, символьний, тд.).**

```python
import numpy as np

# Визначаємо структуру типу даних для масиву:
# 'name' — текстова колонка (тип Unicode, довжина до 20 символів),
# 'mark' — числова колонка (тип int8, цілі числа в діапазоні -128 до 127).
dt = np.dtype([('name', 'U20'), ('mark', np.int8)])  # Замінили np.compat.unicode на 'U20'

# Дані для масиву: кожен елемент містить ім'я (name) і оцінку (mark).
arr = [
    ("Lebiga", 5), 
    ("Thetremba", 5),
    ("S.T.A.L.K.E.R.", 5),
]

# Створюємо масив з указаною структурою типу даних.
C = np.array(arr, dtype=dt)

# Виводимо масив у текстовому поданні.
print(f"Отриманий масив має вигляд:\n{C}")

# Звертаємося до значень колонки 'mark' за її назвою.
print(f"Доступитись до певної колонки тепер можна за її іменем: {C['mark']}")

homework("Як вивести всі імена присутні в даному масиві?")
print(f"Всі імена присутні в даному масиві: {C['name']}")

```

### **Запис та зчитування з файлу, формат CSV**

- `savez` та `load` найпростіші функції для збереження даних `NPZ` формат;
- `CSV` - формат, де значення відділені комою. Деколи цей формат дуже подібний до комірок Excel; 
- `savetxt` - метод дозволяє збуурегти масив у файл; 
- `genfromtxt` - метод вичитує дані з файлі;

```python
import numpy as np
import os

# Оновлений dtype з додаванням нового поля 'group'
dt = np.dtype([('name', np.str_, 20), ('mark', np.int8), ('group', np.str_, 10)])

# Початковий масив, додаємо нове поле 'group'
arr = [
    ("Bohdan", 4, "КІ-21"),
    ("Marta", 5, "КН-31"),
    ("Noname", 0, "ДТ-13"),
    ("Саша", 5, "КН-41")
]

# Створення масиву з заданим типом даних
C = np.array(arr, dtype=dt)

# Збереження масиву у файл .npz
np.savez("my_mass.npz", my_mass=C)
D = np.load("my_mass.npz")
print("Прочитаний з файлу: ", D["my_mass"])

# Збереження масиву у CSV файл
filename = "temp.csv"
np.savetxt(filename, C, fmt="%s,%d,%s", header="name,mark,group", delimiter=",")

# Перевірка чи файл створився
print(f"Перевіряємо чи файл {filename} створився: {os.listdir()}")

# Читання CSV файлу
print("Читаємо файл за допомогою оператора with та методу readlines")
with open(filename) as f:
    D = f.readlines()

print(f"Прочитаний файл:\n {D} \n- як бачимо - {len(D)} елементи є стрічками {type(D[0])}.")

# Зчитування значень з файлу у масив
D = np.genfromtxt(filename, dtype=dt, delimiter=",", names=True)
print(f"Зчитане значення з файла:\n {D}, як бачимо - {D.size} елементи це {type(D[0])} \n- вбудований клас бібліотеки numpy.")
print(f"Доступитись до певного елемента можна наступним чином, наприклад імя першого елементу: {D[0]['name']}")

# Читання вказаного імені з масиву
name_find = D[D['name'] == "Саша"]['name']
print(f"Ім'я, знайдене у масиві: {name_find[0]}")  # Виведення результату
```
