___
### üîµ–£–º–æ–≤–Ω—ñ –û–ø–µ—Ä–∞—Ç–æ—Ä–∏üîµ
–£–º–æ–≤–Ω—ñ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∏ –¥–æ–∑–≤–æ–ª—è—é—Ç—å –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ —Ä—ñ–∑–Ω—ñ –¥—ñ—ó –Ω–∞ –æ—Å–Ω–æ–≤—ñ —É–º–æ–≤. –û—Å–Ω–æ–≤–Ω—ñ –∫–ª—é—á–æ–≤—ñ —Å–ª–æ–≤–∞:

- **if**: –í–∏–∫–æ–Ω—É—î –∫–æ–¥, —è–∫—â–æ —É–º–æ–≤–∞ —ñ—Å—Ç–∏–Ω–Ω–∞.

- **elif**: –ü–µ—Ä–µ–≤—ñ—Ä—è—î –¥–æ–¥–∞—Ç–∫–æ–≤—ñ —É–º–æ–≤–∏, —è–∫—â–æ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ —É–º–æ–≤–∏ –±—É–ª–∏ —Ö–∏–±–Ω–∏–º–∏.

- **else**: –í–∏–∫–æ–Ω—É—î –∫–æ–¥, —è–∫—â–æ –≤—Å—ñ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ —É–º–æ–≤–∏ —Ö–∏–±–Ω—ñ.

`````python
age = 20

if age < 18:
    print("–í–∞–º –º–µ–Ω—à–µ 18 —Ä–æ–∫—ñ–≤")
elif age == 18:
    print("–í–∞–º 18 —Ä–æ–∫—ñ–≤")
else:
    print("–í–∞–º –±—ñ–ª—å—à–µ 18 —Ä–æ–∫—ñ–≤")
`````

___
### üîµ–û–ø–µ—Ä–∞—Ç–æ—Ä–∏ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è —Ç–∞ –ª–æ–≥—ñ—á–Ω—ñ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∏:üîµ
- **–û–ø–µ—Ä–∞—Ç–æ—Ä–∏ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è**: == (–¥–æ—Ä—ñ–≤–Ω—é—î), != (–Ω–µ –¥–æ—Ä—ñ–≤–Ω—é—î), <, >, <=, >=.
- **–õ–æ–≥—ñ—á–Ω—ñ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∏**: and, or, not –¥–ª—è –∫–æ–º–±—ñ–Ω—É–≤–∞–Ω–Ω—è —É–º–æ–≤.

`````python
temperature = 17
is_sunny = True

if temperature < 20 and is_sunny:
    print("–ß—É–¥–æ–≤–∏–π –¥–µ–Ω—å!")
else:
    print("–ù–µ –Ω–∞–π–∫—Ä–∞—â–∏–π –¥–µ–Ω—å.")
`````

___
### üîµ**–¶–∏–∫–ª–∏**üîµ
–¶–∏–∫–ª–∏ –¥–æ–∑–≤–æ–ª—è—é—Ç—å –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ –ø–æ–≤—Ç–æ—Ä—é–≤–∞–Ω—ñ –¥—ñ—ó. Python –ø—ñ–¥—Ç—Ä–∏–º—É—î for —Ç–∞ while —Ü–∏–∫–ª–∏.

**–¶–∏–∫–ª for** - –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è —ñ—Ç–µ—Ä–∞—Ü—ñ—ó –ø–æ –∫–æ–ª–µ–∫—Ü—ñ—è–º (—Å–ø–∏—Å–∫–∞–º, —Ä—è–¥–∫–∞–º, —Å–ª–æ–≤–Ω–∏–∫–∞–º).
- –§—É–Ω–∫—Ü—ñ—è **range** –≥–µ–Ω–µ—Ä—É—î –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å —á–∏—Å–µ–ª.

**–¶–∏–∫–ª while** –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è, –ø–æ–∫–∏ —É–º–æ–≤–∞ —î —ñ—Å—Ç–∏–Ω–Ω–æ—é. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è, –∫–æ–ª–∏ –º–∏ –Ω–µ –∑–Ω–∞—î–º–æ —Ç–æ—á–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –ø–æ–≤—Ç–æ—Ä–µ–Ω—å.

#### –í–∏—Ö—ñ–¥ –∑ —Ü–∏–∫–ª—É: break —Ç–∞ continue
- break: –ü–µ—Ä–µ—Ä–∏–≤–∞—î —Ü–∏–∫–ª —ñ –≤–∏—Ö–æ–¥–∏—Ç—å –∑ –Ω—å–æ–≥–æ.
- continue: –ü—Ä–æ–ø—É—Å–∫–∞—î –ø–æ—Ç–æ—á–Ω—É —ñ—Ç–µ—Ä–∞—Ü—ñ—é —Ç–∞ –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç—å –¥–æ –Ω–∞—Å—Ç—É–ø–Ω–æ—ó.

`````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("I like", fruit)
print()

for i in range(3):  # –í–∏–≤–µ–¥–µ —á–∏—Å–ª–∞ –≤—ñ–¥ 0 –¥–æ 2
    print(i)
print()

for i in range(1, 4):  # –í–∏–≤–µ–¥–µ —á–∏—Å–ª–∞ –≤—ñ–¥ 1 –¥–æ 3
    print(i)
print()

count = 0
while count < 3:
    print("Count is", count)
    count += 1  
print()

for i in range(10):
    if i == 5:
        break  # –¶–∏–∫–ª –∑–∞–≤–µ—Ä—à–∏—Ç—å—Å—è, –∫–æ–ª–∏ i –¥–æ—Ä—ñ–≤–Ω—é—î 5
    print(i)

for i in range(10):
    if i % 2 == 0:
        continue  # –ü—Ä–æ–ø—É—Å–∫–∞—î –ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞
    print(i)  # –í–∏–≤–µ–¥–µ –ª–∏—à–µ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞
`````

### üîµ–û—Å–Ω–æ–≤–Ω—ñ –ö–æ–ª–µ–∫—Ü—ñ—ó Pythonüîµ

1. #### ‚ú≥Ô∏è **–°–ø–∏—Å–æ–∫ (List)** ‚ú≥Ô∏è


**List** ‚Äì —Ü–µ –∑–º—ñ–Ω–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è, —è–∫–∞ –¥–æ–∑–≤–æ–ª—è—î –∑–±–µ—Ä—ñ–≥–∞—Ç–∏ –∫—ñ–ª—å–∫–∞ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —É –ø–µ–≤–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É. –í–∏ –º–æ–∂–µ—Ç–µ –∑–º—ñ–Ω—é–≤–∞—Ç–∏ —Å–ø–∏—Å–æ–∫, –¥–æ–¥–∞—é—á–∏ –∞–±–æ –≤–∏–¥–∞–ª—è—é—á–∏ –µ–ª–µ–º–µ–Ω—Ç–∏. –°–ø–∏—Å–∫–∏ –ø—ñ–¥—Ç—Ä–∏–º—É—é—Ç—å —ñ–Ω–¥–µ–∫—Å–∞—Ü—ñ—é, —Ç–æ–±—Ç–æ –¥–æ—Å—Ç—É–ø –¥–æ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∑–∞ —ó—Ö –ø–æ—Ä—è–¥–∫–æ–≤–∏–º –Ω–æ–º–µ—Ä–æ–º (–ø–æ—á–∏–Ω–∞—é—á–∏ –∑ –Ω—É–ª—è).

`````python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É
fruits = ["apple", "banana", "cherry"]

# –î–æ—Å—Ç—É–ø –¥–æ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
print(fruits[0])  # –í–∏–≤–µ–¥–µ 'apple'
print(fruits[1])  # –í–∏–≤–µ–¥–µ 'banana'

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞
fruits.append("orange")
print(fruits)  # –í–∏–≤–µ–¥–µ ['apple', 'banana', 'cherry', 'orange']

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞
fruits.remove("banana")
print(fruits)  # –í–∏–≤–µ–¥–µ ['apple', 'cherry', 'orange']
`````

2. #### ‚ú≥Ô∏è **–ö–æ—Ä—Ç–µ–∂ (Tuple)** ‚ú≥Ô∏è

**Tuple** ‚Äì —Ü–µ –Ω–µ–∑–º—ñ–Ω–Ω–∞ –∫–æ–ª–µ–∫—Ü—ñ—è, —Å—Ö–æ–∂–∞ –Ω–∞ —Å–ø–∏—Å–æ–∫, –∞–ª–µ –ø—ñ—Å–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∏ –≤ –∫–æ—Ä—Ç–µ–∂—ñ –Ω–µ –º–æ–∂–Ω–∞ –∑–º—ñ–Ω—é–≤–∞—Ç–∏ (–¥–æ–¥–∞–≤–∞—Ç–∏ —á–∏ –≤–∏–¥–∞–ª—è—Ç–∏). –ö–æ—Ä—Ç–µ–∂—ñ —á–∞—Å—Ç–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö, —è–∫—ñ –Ω–µ –ø–æ–≤–∏–Ω–Ω—ñ –∑–º—ñ–Ω—é–≤–∞—Ç–∏—Å—è.

`````python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –∫–æ—Ä—Ç–µ–∂—É
coordinates = (10.0, 20.5)

# –î–æ—Å—Ç—É–ø –¥–æ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤
print(coordinates[0])  # –í–∏–≤–µ–¥–µ 10.0
print(coordinates[1])  # –í–∏–≤–µ–¥–µ 20.5

# –°–ø—Ä–æ–±–∞ –∑–º—ñ–Ω–∏—Ç–∏ –∫–æ—Ä—Ç–µ–∂ –≤–∏–∫–ª–∏—á–µ –ø–æ–º–∏–ª–∫—É:
# coordinates[0] = 15.0  # –¶–µ –ø—Ä–∏–∑–≤–µ–¥–µ –¥–æ TypeError, –æ—Å–∫—ñ–ª—å–∫–∏ –∫–æ—Ä—Ç–µ–∂ —î –Ω–µ–∑–º—ñ–Ω–Ω–∏–º
`````

3. #### ‚ú≥Ô∏è **–ú–Ω–æ–∂–∏–Ω–∞ (Set)** ‚ú≥Ô∏è

**Set** ‚Äì —Ü–µ –∫–æ–ª–µ–∫—Ü—ñ—è, —è–∫–∞ –Ω–µ –∑–±–µ—Ä—ñ–≥–∞—î –µ–ª–µ–º–µ–Ω—Ç–∏ –≤ –ø–µ–≤–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É —Ç–∞ –Ω–µ –¥–æ–ø—É—Å–∫–∞—î –¥—É–±–ª—é–≤–∞–Ω–Ω—è. –ú–Ω–æ–∂–∏–Ω–∏ –∫–æ—Ä–∏—Å–Ω—ñ, –∫–æ–ª–∏ –ø–æ—Ç—Ä—ñ–±–Ω–æ –∑–±–µ—Ä—ñ–≥–∞—Ç–∏ —É–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∞–±–æ –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ –æ–ø–µ—Ä–∞—Ü—ñ—ó –æ–±'—î–¥–Ω–∞–Ω–Ω—è —Ç–∞ –ø–µ—Ä–µ—Ç–∏–Ω—É.

`````python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–Ω–æ–∂–∏–Ω–∏
unique_numbers = {1, 2, 3, 4}

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞
unique_numbers.add(5)
print(unique_numbers)  # –í–∏–≤–µ–¥–µ, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, {1, 2, 3, 4, 5}

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç–∞
unique_numbers.remove(3)
print(unique_numbers)  # –í–∏–≤–µ–¥–µ, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, {1, 2, 4, 5}

# –û–ø–µ—Ä–∞—Ü—ñ—ó –Ω–∞–¥ –º–Ω–æ–∂–∏–Ω–∞–º–∏
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 | set2)  # –û–±'—î–¥–Ω–∞–Ω–Ω—è, –≤–∏–≤–µ–¥–µ {1, 2, 3, 4, 5}
print(set1 & set2)  # –ü–µ—Ä–µ—Ç–∏–Ω, –≤–∏–≤–µ–¥–µ {3}
`````

4. #### ‚ú≥Ô∏è **–°–ª–æ–≤–Ω–∏–∫ (Dictionary)** ‚ú≥Ô∏è

**Dictionary** ‚Äì —Ü–µ –∫–æ–ª–µ–∫—Ü—ñ—è –ø–∞—Ä –∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–Ω—è, –¥–µ –∫–æ–∂–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è –º–∞—î —Å–≤—ñ–π —É–Ω—ñ–∫–∞–ª—å–Ω–∏–π –∫–ª—é—á. –ö–ª—é—á—ñ –≤ —Å–ª–æ–≤–Ω–∏–∫—É –ø–æ–≤–∏–Ω–Ω—ñ –±—É—Ç–∏ —É–Ω—ñ–∫–∞–ª—å–Ω–∏–º–∏ —Ç–∞ –Ω–µ–∑–º—ñ–Ω–Ω–∏–º–∏ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —Ä—è–¥–∫–∏ –∞–±–æ —á–∏—Å–ª–∞). –°–ª–æ–≤–Ω–∏–∫–∏ –¥—É–∂–µ –∫–æ—Ä–∏—Å–Ω—ñ –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑—ñ –∑–≤‚Äô—è–∑–∫–∞–º–∏, —Ç–∞–∫–∏—Ö —è–∫ —ñ–º–µ–Ω–∞ —Ç–∞ —ó—Ö–Ω—ñ–π –≤—ñ–∫.

`````python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ª–æ–≤–Ω–∏–∫–∞
person = {"name": "Ash", "age": 38, "city": "Kyiv"}

# –î–æ—Å—Ç—É–ø –¥–æ –∑–Ω–∞—á–µ–Ω—å
print(person["name"])  # –í–∏–≤–µ–¥–µ 'Ash'
print(person["age"])   # –í–∏–≤–µ–¥–µ 38

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –Ω–æ–≤–æ—ó –ø–∞—Ä–∏ –∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–Ω—è
person["email"] = "ash@example.com"
print(person)

# –í–∏–¥–∞–ª–µ–Ω–Ω—è –ø–∞—Ä–∏ –∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–Ω—è
del person["age"]
print(person)  # –í–∏–≤–µ–¥–µ {'name': 'Ash', 'city': 'Kyiv', 'email': 'ash@example.com'}
`````

### üîµ **–†—è–¥–æ–∫** (—Ç–∏–ø `str`) üîµ

 –¶–µ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å —Å–∏–º–≤–æ–ª—ñ–≤, —è–∫–∞ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –∑–±–µ—Ä—ñ–≥–∞–Ω–Ω—è —Ç–µ–∫—Å—Ç—É. –†—è–¥–∫–∏ –æ–≥–æ—Ä—Ç–∞—é—Ç—å—Å—è –≤ –æ–¥–∏–Ω–∞—Ä–Ω—ñ (`'...'`) –∞–±–æ –ø–æ–¥–≤—ñ–π–Ω—ñ –ª–∞–ø–∫–∏ (`"..."`) —ñ —î –Ω–µ–∑–º—ñ–Ω–Ω–∏–º–∏, —Ç–æ–±—Ç–æ –Ω–µ –º–æ–∂—É—Ç—å –±—É—Ç–∏ –∑–º—ñ–Ω–µ–Ω—ñ –ø—ñ—Å–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è. –î–µ—è–∫—ñ –æ—Å–Ω–æ–≤–Ω—ñ –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ –π –º–µ—Ç–æ–¥–∏ —Ä–æ–±–æ—Ç–∏ –∑ —Ä—è–¥–∫–∞–º–∏:

#### ‚ú≥Ô∏è –û—Å–Ω–æ–≤–Ω—ñ –í–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ:

3. **–ö–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü—ñ—è**: –æ–±'—î–¥–Ω–∞–Ω–Ω—è —Ä—è–¥–∫—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ `+`.
4. **–ü–æ–≤—Ç–æ—Ä–µ–Ω–Ω—è**: —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥–∫–∞, —è–∫–∏–π –ø–æ–≤—Ç–æ—Ä—é—î—Ç—å—Å—è –∫—ñ–ª—å–∫–∞ —Ä–∞–∑—ñ–≤, –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `*`.

#### ‚ú≥Ô∏è –û—Å–Ω–æ–≤–Ω—ñ –ú–µ—Ç–æ–¥–∏:
- `len(—Å—Ç—Ä–æ–∫)`: –ø–æ–≤–µ—Ä—Ç–∞—î –¥–æ–≤–∂–∏–Ω—É —Ä—è–¥–∫–∞.
- `str.lower()` —ñ `str.upper()`: –∑–º—ñ–Ω—é—é—Ç—å —Ä–µ–≥—ñ—Å—Ç—Ä —Å–∏–º–≤–æ–ª—ñ–≤.
- `str.strip()`: –≤–∏–¥–∞–ª—è—î –ø—Ä–æ–±—ñ–ª–∏ –∑ –ø–æ—á–∞—Ç–∫—É —ñ –∫—ñ–Ω—Ü—è —Ä—è–¥–∫–∞.
- `str.replace("—Å—Ç–∞—Ä–∏–π", "–Ω–æ–≤–∏–π")`: –∑–∞–º—ñ–Ω—é—î –ø—ñ–¥—Ä—è–¥–æ–∫ –Ω–∞ —ñ–Ω—à–∏–π.
- `str.find("–ø—ñ–¥—Ä—è–¥–æ–∫")`: –ø–æ–≤–µ—Ä—Ç–∞—î —ñ–Ω–¥–µ–∫—Å –ø–µ—Ä—à–æ–≥–æ –≤—Ö–æ–¥–∂–µ–Ω–Ω—è –ø—ñ–¥—Ä—è–¥–∫–∞ –∞–±–æ `-1`, —è–∫—â–æ –π–æ–≥–æ –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ.

### üîµ –§–æ—Ä–º–∞—Ç–æ–≤–∞–Ω—ñ —Ä—è–¥–∫–∏ (f-strings) üîµ

**–û—Å–Ω–æ–≤–Ω—ñ –û—Å–æ–±–ª–∏–≤–æ—Å—Ç—ñ:**
- –ü–æ–∑–Ω–∞—á–∞—é—Ç—å—Å—è –ª—ñ—Ç–µ—Ä–æ—é f –ø–µ—Ä–µ–¥ –ª–∞–ø–∫–∞–º–∏: f"..." –∞–±–æ f'...'.
- –í—Å—Ç–∞–≤–∫–∞ –∑–º—ñ–Ω–Ω–∏—Ö —Ç–∞ –≤–∏—Ä–∞–∑—ñ–≤: –∑–Ω–∞—á–µ–Ω–Ω—è –º–æ–∂–Ω–∞ –≤—Å—Ç–∞–≤–ª—è—Ç–∏ –±–µ–∑–ø–æ—Å–µ—Ä–µ–¥–Ω—å–æ –≤ —Ç–µ–∫—Å—Ç –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é {}.

`````python

letters = ["a", "b", "c"]
for i in range(len(letters)):
    print(f"–ù–∞ –ø–æ–∑–∏—Ü—ñ—ó {i} –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –±—É–∫–≤–∞ {letters[i]}")

# –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª –∑ –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è–º
value = 123.456789
print(f"–ó–Ω–∞—á–µ–Ω–Ω—è –∑ –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è–º –¥–æ —Ç—Ä—å–æ—Ö –∑–Ω–∞–∫—ñ–≤: {value:.3f}")  

# –í–∏—Ä—ñ–≤–Ω—é–≤–∞–Ω–Ω—è —Ç–µ–∫—Å—Ç—É —Ç–∞ —á–∏—Å–µ–ª
print(f"|{'left':<10}|{'center':^10}|{'right':>10}|")  
`````

`elif` (optional) ‚Äì –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞ —É–º–æ–≤–∞, —â–æ –ø–µ—Ä–µ–≤—ñ—Ä—è—î—Ç—å—Å—è, —è–∫—â–æ –ø–æ–ø–µ—Ä–µ–¥–Ω—è —É–º–æ–≤–∞ —î —Ö–∏–±–Ω–æ—é (False).
`else` (optional) ‚Äì –≤–∏–∫–æ–Ω—É—î –±–ª–æ–∫ –∫–æ–¥—É, —è–∫—â–æ –∂–æ–¥–Ω–∞ –∑ —É–º–æ–≤ –Ω–µ –≤–∏–∫–æ–Ω–∞–ª–∞—Å—è.

#### ‚ú≥Ô∏è –£–º–æ–≤–Ω—ñ –í–∏—Ä–∞–∑–∏:
–£–º–æ–≤–∏ –º–æ–∂—É—Ç—å –ø–æ—Ä—ñ–≤–Ω—é–≤–∞—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –æ–ø–µ—Ä–∞—Ç–æ—Ä—ñ–≤, —Ç–∞–∫–∏—Ö —è–∫ `==`, `!=`, `<`, `>`, `<=`, `>=`.
–õ–æ–≥—ñ—á–Ω—ñ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∏ `and`, `or`, `not` –¥–æ–∑–≤–æ–ª—è—é—Ç—å –∫–æ–º–±—ñ–Ω—É–≤–∞—Ç–∏ –∫—ñ–ª—å–∫–∞ —É–º–æ–≤.

`````python
temperature = 16

if temperature > 22:
    print("It's hot outside.")
elif temperature > 13:
    print("It's warm outside.")
else:
    print("It's cold outside.")



age = 22
is_student = True
purchase_amount = 120

if age < 18:
    print("–í–∏ –æ—Ç—Ä–∏–º–∞—î—Ç–µ –∑–Ω–∏–∂–∫—É 20% –¥–ª—è –Ω–µ–ø–æ–≤–Ω–æ–ª—ñ—Ç–Ω—ñ—Ö.")
elif age >= 18 and age <= 25:
    if is_student and purchase_amount > 100:
        print("–í–∏ –æ—Ç—Ä–∏–º–∞—î—Ç–µ –∑–Ω–∏–∂–∫—É 15% –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤ —ñ–∑ –ø–æ–∫—É–ø–∫–æ—é –ø–æ–Ω–∞–¥ 100.")
    elif is_student and purchase_amount <= 100:
        print("–í–∏ –æ—Ç—Ä–∏–º–∞—î—Ç–µ –∑–Ω–∏–∂–∫—É 10% –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤.")
    else:
        print("–í–∏ –æ—Ç—Ä–∏–º–∞—î—Ç–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—É –∑–Ω–∏–∂–∫—É 5%.")
elif age > 25 and purchase_amount > 200:
    print("–í–∏ –æ—Ç—Ä–∏–º–∞—î—Ç–µ –∑–Ω–∏–∂–∫—É 10% –¥–ª—è –ø–æ–∫—É–ø–æ–∫ –ø–æ–Ω–∞–¥ 200.")
else:
    print("–ó–Ω–∏–∂–∫–∞ –Ω–µ –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–∞.")
`````

### üîµ **–ö–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—è try-except-finally** üîµ

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –æ–±—Ä–æ–±–∫–∏ –≤–∏–Ω—è—Ç–∫—ñ–≤ (–ø–æ–º–∏–ª–æ–∫), —â–æ –º–æ–∂—É—Ç—å –≤–∏–Ω–∏–∫–Ω—É—Ç–∏ –ø—ñ–¥ —á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–¥—É. –¶–µ –¥–æ–∑–≤–æ–ª—è—î —É–Ω–∏–∫–Ω—É—Ç–∏ –ø–æ–º–∏–ª–æ–∫, —è–∫—ñ –º–æ–≥–ª–∏ –± –ø–µ—Ä–µ—Ä–≤–∞—Ç–∏ —Ä–æ–±–æ—Ç—É –ø—Ä–æ–≥—Ä–∞–º–∏, —Ç–∞ –≤–∏–∫–æ–Ω–∞—Ç–∏ –ø–µ–≤–Ω–∏–π –∫–æ–¥ –Ω–µ–∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ —Ç–æ–≥–æ, —Å—Ç–∞–ª–∞—Å—è –ø–æ–º–∏–ª–∫–∞ —á–∏ –Ω—ñ.

#### ‚ú≥Ô∏è –û—Å–Ω–æ–≤–Ω—ñ –ö–æ–º–ø–æ–Ω–µ–Ω—Ç–∏:
- `try` ‚Äì –±–ª–æ–∫, —É —è–∫–æ–º—É –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∫–æ–¥, —â–æ –ø–æ—Ç–µ–Ω—Ü—ñ–π–Ω–æ –º–æ–∂–µ –≤–∏–∫–ª–∏–∫–∞—Ç–∏ –ø–æ–º–∏–ª–∫—É.
- `except` ‚Äì –±–ª–æ–∫, —É —è–∫–æ–º—É –æ–±—Ä–æ–±–ª—è—î—Ç—å—Å—è –ø–æ–º–∏–ª–∫–∞, —è–∫—â–æ –≤–æ–Ω–∞ –≤–∏–Ω–∏–∫–ª–∞. –ú–æ–∂–Ω–∞ –≤–∫–∞–∑–∞—Ç–∏ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–∏–π —Ç–∏–ø –≤–∏–Ω—è—Ç–∫–∞, —â–æ–± –æ–±—Ä–æ–±–ª—è—Ç–∏ —Ç—ñ–ª—å–∫–∏ –ø–µ–≤–Ω—ñ —Ç–∏–ø–∏ –ø–æ–º–∏–ª–æ–∫.
- `finally` ‚Äì –±–ª–æ–∫, —è–∫–∏–π –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∑–∞–≤–∂–¥–∏, –Ω–µ–∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ —Ç–æ–≥–æ, –≤–∏–Ω–∏–∫–ª–∞ –ø–æ–º–∏–ª–∫–∞ —á–∏ –Ω—ñ. –ô–æ–≥–æ –∑–∞–∑–≤–∏—á–∞–π –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å –¥–ª—è "–æ—á–∏—â–µ–Ω–Ω—è" (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –∑–∞–∫—Ä–∏—Ç—Ç—è —Ñ–∞–π–ª—ñ–≤ –∞–±–æ –∑–≤—ñ–ª—å–Ω–µ–Ω–Ω—è —Ä–µ—Å—É—Ä—Å—ñ–≤).

`````python
A = 0
try:
    print("–©–æ –±—É–¥–µ —è–∫—â–æ", 10/A, "?")
except Exception as e:
    print(e)
finally:
    print("–ê –≤–æ—Ç –≤–æ–Ω–æ —â–æ!")
`````

`````python
try:
    number = 5  # –ó–∞–º—ñ–Ω–∞ –Ω–∞ –∑–∞–¥–∞–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è, –∑–∞–º—ñ—Å—Ç—å –≤–≤–æ–¥—É –∑ –∫–ª–∞–≤—ñ–∞—Ç—É—Ä–∏
    result = 10 / number
    print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", result)
except ValueError:
    print("–ü–æ–º–∏–ª–∫–∞: –≤–≤–µ–¥–µ–Ω–æ –Ω–µ—á–∏—Å–ª–æ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è!")
except ZeroDivisionError:
    print("–ü–æ–º–∏–ª–∫–∞: –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ –Ω—É–ª—å!")
finally:
    print("–¶–µ–π —Ä—è–¥–æ–∫ –±—É–¥–µ –Ω–∞–¥—Ä—É–∫–æ–≤–∞–Ω–∏–π –∑–∞–≤–∂–¥–∏.")
`````

### üîµ**–ö–æ–Ω—Ç–µ–∫—Å—Ç-–º–µ–Ω–µ–¥–∂–µ—Ä `with`** üîµ

**–û—Å–Ω–æ–≤–Ω—ñ –ü–µ—Ä–µ–≤–∞–≥–∏:**
- —Ä–µ—Å—É—Ä—Å –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –∑–∞–∫—Ä–∏—î—Ç—å—Å—è –ø—ñ—Å–ª—è –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –±–ª–æ–∫—É –∫–æ–¥—É with, —â–æ –∑–∞–ø–æ–±—ñ–≥–∞—î –≤–∏—Ç–æ–∫—É –ø–∞–º'—è—Ç—ñ.
- –∫–æ–¥ —ñ–∑ with –≤–∏–≥–ª—è–¥–∞—î —á–∏—Å—Ç—ñ—à–∏–º —ñ –ø—Ä–æ—Å—Ç—ñ—à–∏–º.

#### ‚ú≥Ô∏è **–Ø–∫ –ø—Ä–∞—Ü—é—î with:**
1. –í—Ö—ñ–¥ —É –±–ª–æ–∫ with: –ö–æ–ª–∏ Python –¥–æ—Å—è–≥–∞—î —Ä—è–¥–∫–∞ –∑ with, –≤—ñ–Ω –≤–∏–∫–ª–∏–∫–∞—î –º–µ—Ç–æ–¥ __enter__() –æ–±‚Äô—î–∫—Ç–∞, —è–∫–∏–π –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è —è–∫ –∫–æ–Ω—Ç–µ–∫—Å—Ç-–º–µ–Ω–µ–¥–∂–µ—Ä.

1. –ú–µ—Ç–æ–¥ __enter__() —á–∞—Å—Ç–æ –≤—ñ–¥–∫—Ä–∏–≤–∞—î –∞–±–æ —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—î —Ä–µ—Å—É—Ä—Å (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —Ñ–∞–π–ª) —ñ –º–æ–∂–µ –ø–æ–≤–µ—Ä–Ω—É—Ç–∏ —Ä–µ—Å—É—Ä—Å (–∞–±–æ —ñ–Ω—à–∏–π –æ–±'—î–∫—Ç), —è–∫–∏–π –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è —É –±–ª–æ—Ü—ñ with.
–í–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–¥—É –≤ –±–ª–æ—Ü—ñ: –ö–æ–¥ —É –±–ª–æ—Ü—ñ with (–≤—ñ–¥—Å—Ç—É–ø–ª–µ–Ω–∏–π –ø—ñ–¥ with) –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ü—å–æ–≥–æ —Ä–µ—Å—É—Ä—Å—É.

1. –í–∏—Ö—ñ–¥ –∑ –±–ª–æ–∫—É with: –ü—ñ—Å–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—è –±–ª–æ–∫—É with (–Ω–µ–∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ —Ç–æ–≥–æ, —á–∏ –∑–∞–∫—ñ–Ω—á–∏–≤—Å—è –≤—ñ–Ω –Ω–æ—Ä–º–∞–ª—å–Ω–æ, —á–∏ –∑ –ø–æ–º–∏–ª–∫–æ—é), Python –≤–∏–∫–ª–∏–∫–∞—î –º–µ—Ç–æ–¥ __exit__() –æ–±'—î–∫—Ç–∞.

1. –ú–µ—Ç–æ–¥ __exit__() –∑–∞–∑–≤–∏—á–∞–π –∑–∞–∫—Ä–∏–≤–∞—î –∞–±–æ –∑–≤—ñ–ª—å–Ω—è—î —Ä–µ—Å—É—Ä—Å, —â–æ –¥–æ–∑–≤–æ–ª—è—î –∑–µ–∫–æ–Ω–æ–º–∏—Ç–∏ –ø–∞–º'—è—Ç—å —ñ —É–Ω–∏–∫–Ω—É—Ç–∏ –≤–∏—Ç–æ–∫—ñ–≤ —Ä–µ—Å—É—Ä—Å—ñ–≤.
–Ø–∫—â–æ —É –±–ª–æ—Ü—ñ with –≤–∏–Ω–∏–∫–ª–∞ –ø–æ–º–∏–ª–∫–∞, __exit__() —Ç–∞–∫–æ–∂ –æ–±—Ä–æ–±–∏—Ç—å —ó—ó (—è–∫—â–æ —Ü–µ –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–æ) –∞–±–æ –¥–æ–∑–≤–æ–ª–∏—Ç—å –ø–æ–º–∏–ª—Ü—ñ –ø—ñ–¥–Ω—è—Ç–∏—Å—è –¥–∞–ª—ñ.

`````python
class MyResource:
    def __enter__(self):
        print("–†–µ—Å—É—Ä—Å –≤—ñ–¥–∫—Ä–∏—Ç–æ")
        return self  # –ü–æ–≤–µ—Ä—Ç–∞—î –æ–±'—î–∫—Ç –¥–ª—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –≤ –±–ª–æ—Ü—ñ with

    def __exit__(self, exc_type, exc_value, traceback):
        print("–†–µ—Å—É—Ä—Å –∑–∞–∫—Ä–∏—Ç–æ")

# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –≤–ª–∞—Å–Ω–æ–≥–æ –∫–æ–Ω—Ç–µ–∫—Å—Ç-–º–µ–Ω–µ–¥–∂–µ—Ä–∞
with MyResource() as resource:
    print("–í–∏–∫–æ–Ω–∞–Ω–Ω—è –æ–ø–µ—Ä–∞—Ü—ñ–π —ñ–∑ —Ä–µ—Å—É—Ä—Å–æ–º")
`````

### üîµ **lambda-—Ñ—É–Ω–∫—Ü—ñ—ó –≤ Python** üîµ

–¶–µ –∞–Ω–æ–Ω—ñ–º–Ω—ñ, –∫–æ—Ä–æ—Ç–∫—ñ —Ñ—É–Ω–∫—Ü—ñ—ó, —è–∫—ñ –∑–∞–∑–≤–∏—á–∞–π —Å—Ç–≤–æ—Ä—é—é—Ç—å—Å—è –¥–ª—è –ø—Ä–æ—Å—Ç–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π. –í–æ–Ω–∏ –¥–æ–∑–≤–æ–ª—è—é—Ç—å –ø–∏—Å–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó –≤ –æ–¥–∏–Ω —Ä—è–¥–æ–∫, –±–µ–∑ –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ—Å—Ç—ñ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –∫–ª—é—á–æ–≤–µ —Å–ª–æ–≤–æ def —ñ —è–≤–Ω–æ –∑–∞–¥–∞–≤–∞—Ç–∏ —ñ–º'—è —Ñ—É–Ω–∫—Ü—ñ—ó. lambda-—Ñ—É–Ω–∫—Ü—ñ—ó –æ—Å–æ–±–ª–∏–≤–æ –∑—Ä—É—á–Ω—ñ, –∫–æ–ª–∏ —Ñ—É–Ω–∫—Ü—ñ—é –ø–æ—Ç—Ä—ñ–±–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç–∏ —è–∫ –∞—Ä–≥—É–º–µ–Ω—Ç —ñ–Ω—à–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∞–±–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–≤–∞—Ç–∏ –æ–¥–Ω–æ—Ä–∞–∑–æ–≤–æ.

#### ‚ú≥Ô∏è –°–∏–Ω—Ç–∞–∫—Å–∏—Å: `lambda` –∞—Ä–≥—É–º–µ–Ω—Ç–∏: –≤–∏—Ä–∞–∑
- –∞—Ä–≥—É–º–µ–Ω—Ç–∏: –ø–µ—Ä–µ–ª—ñ–∫ –≤—Ö—ñ–¥–Ω–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤, —è–∫—ñ —Ñ—É–Ω–∫—Ü—ñ—è –ø—Ä–∏–π–º–∞—î.
- –≤–∏—Ä–∞–∑: –æ–¥–∏–Ω —Ä—è–¥–æ–∫ –∫–æ–¥—É, —è–∫–∏–π –ø–æ–≤–µ—Ä—Ç–∞—î –∑–Ω–∞—á–µ–Ω–Ω—è. –ü—ñ—Å–ª—è –Ω—å–æ–≥–æ –Ω–µ–º–∞—î –Ω–µ–æ–±—Ö—ñ–¥–Ω–æ—Å—Ç—ñ –ø–∏—Å–∞—Ç–∏ return, –æ—Å–∫—ñ–ª—å–∫–∏ lambda –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –ø–æ–≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –≤–∏—Ä–∞–∑—É.


`````python
# –ü—Ä–∏–∫–ª–∞–¥ 1
square = lambda x: x ** 2
print(square(5))  # –í–∏–≤–µ–¥–µ: 25

# –ü—Ä–∏–∫–ª–∞–¥ 2
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)  # –í–∏–≤–µ–¥–µ: [1, 4, 9, 16, 25]

# –ü—Ä–∏–∫–ª–∞–¥ 3
people = [{'name': 'Anna', 'age': 22}, {'name': 'Ben', 'age': 30}]
sorted_people = sorted(people, key=lambda person: person['age'])
print(sorted_people)  # –í–∏–≤–µ–¥–µ: [{'name': 'Anna', 'age': 22}, {'name': 'Ben', 'age': 30}]

# –ü—Ä–∏–∫–ª–∞–¥ 4
this_is_lambda = lambda first, last: f'–¶–µ–π –∫–æ–¥ –≤–∏–≤–æ–¥–∏—Ç—å: {first} {last}'
print("–¶–µ –ø—Ä–æ—Å—Ç–æ —Ñ—É–Ω–∫—Ü—ñ—è:", this_is_lambda)
print("–¶–µ —ó—ó –≤–∏–∫–ª–∏–∫:", this_is_lambda('Jinx', 'was here'))
`````
